# Rangkuman Bab 1 – Konversi ER Diagram ke Skema Relasi

## 1. Tujuan Pembelajaran
- Memahami cara mengubah ERD menjadi skema relasi dan tabel fisik.
- Mempermudah proses transformasi dari desain konseptual ke database nyata.

## 2. Konsep Dasar
### Entitas, Atribut, Primary Key
- Entitas: objek yang datanya disimpan.
- Atribut: karakteristik entitas.
- Primary Key: pembeda unik tiap record.

### Relasi & Kardinalitas
- Relasi: hubungan antar entitas.
- Kardinalitas: 1–1, 1–N, N–M.

## 3. Aturan Konversi ERD ke Skema Relasi
1. **Entitas kuat → tabel.**
2. **Composite atribut → dipecah menjadi atribut sederhana.**
3. **Multivalue atribut → tabel baru.**
4. **Derived atribut → tetap jadi kolom.**
5. **Entitas lemah → tabel dengan PK gabungan + FK.**
6. **Relasi 1–1 → salah satu tabel menyimpan FK.**
7. **Relasi 1–N → sisi N menyimpan FK dari sisi 1.**
8. **Relasi N–M → tabel relasi dengan FK kedua entitas.**
9. **Unary relasi → tabel tetap sama tetapi menambah FK diri sendiri.**
10. **Relasi ternary → tabel relasi dengan 3 FK.**
11. **Generalization/Spesialisasi → 2 metode pembentukan tabel.**
12. **Agregasi → tabel baru yang menampung relasi kompleks.**
